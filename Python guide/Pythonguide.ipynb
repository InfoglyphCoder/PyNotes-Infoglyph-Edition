### **History of Python:**

Python is a computer programming language that was created by a man named **Guido van Rossum** in **the late 1980s**. Guido wanted to make a language that was **easy to read and write**, so people could create computer programs without feeling too stressed about complicated rules.

Before Python, there were other programming languages, but Guido wanted to create something **simple** and **fun**. He worked on it, and in **1991**, Python was officially released for everyone to use. The name **"Python"** was actually inspired by a **fun TV show called "Monty Python's Flying Circus"**, which Guido loved.



#### **Why is Python special?**

1. **Easy to Read and Write**: Python's rules are simple. You don’t need to remember a lot of hard things. It looks a bit like English, which makes it easy to understand.
   
2. **Used Everywhere**: People use Python for all sorts of things—**making games**, **building websites**, and even for **science** and **artificial intelligence** (like robots and smart machines).

3. **Lots of Help**: Python has a huge **community** of people who help each other. There are also many **libraries** (sets of helpful instructions) that make it even easier to do things.

4. **Free to Use**: Python is completely **free**. You don’t have to pay to use it, so anyone can download and start using it right away.


#### **What Happened Over Time?**

- **Python 1.0** was released in **1994**. It already had the basic features like loops, variables, and functions.
  
- **Python 2.0** came out in **2000** with some important changes, like better ways to handle text and numbers.
  
- **Python 3.0**, the version we use today, was released in **2008**. It was **even better**, but it had a lot of changes. Some things from Python 2 don’t work the same way in Python 3, so we usually recommend using Python 3.


#### **Why Do People Like Python?**

- It’s **simple** to start with, like learning how to read and write. You can focus on the fun part of making a program, instead of getting stuck in complicated rules.
  
- It's **powerful**! Even though it's simple, you can make really **cool stuff** like websites, games, and robots.

- **Many people** around the world use Python, and there's a lot of support if you get stuck.

#### **1. Go to Python’s Website**  
First, open your browser (like Chrome). Type **python.org** and press Enter.  

Now you’ll see a big yellow button that says:  
**“Download Python 3.13.1”**  
Click it! That’s Python coming to your computer. 🎉  


---

#### **2. Install Python**  
After it’s done downloading, find the file in your Downloads folder.  
It will look like this: **python-3.13.1.exe**.  

Click on it, and a window will pop up. 



---

#### **3. Don’t Forget the Map! 🗺️**  
There’s a little box that says **“Add Python to PATH”**.  
Click it! This is like giving Python a map so it doesn’t get lost.  

Then, click **Install Now**. Wait for a few moments, and you’re done!  



---

#### **4. Check if Python is Ready**  
Open your **Command Prompt** (like a secret chat box for your computer).  

Type this:  
```
python --version
```  
Press Enter. If it says something like **Python 3.13.1**, hooray! Python is ready to play! 🎉  

---

Now you have Python, your super cool computer helper. Let’s start having fun with it soon! 🐍💻




### ** Get VS Code on Your Computer! 💻**  

VS Code is like a magic notebook where you can write Python and do awesome things. Let’s bring it to your computer step by step.  


---

### **1. Go to VS Code’s Home**  
1. Open your browser (like Chrome or Edge).  
2. Type this in the search bar: **code.visualstudio.com**.  
3. Press Enter, and you’ll see the home page of VS Code.  



---

### **2. Click the Download Button**  
On the page, there’s a big blue button that says:  
**“Download for Windows”** (or Mac, depending on your computer).  
Click it!  

Now your computer will start downloading VS Code



---

### **3. Open the Downloaded File**  
When it’s done downloading, go to your **Downloads** folder.  
You’ll find a file that looks like this: **VSCodeSetup.exe** (or something similar).  

Double-click on it, and it will open the installation wizard. 



---

### **4. Follow the Installation**  
The wizard will guide you step by step. Here’s what you’ll see:  

1. **License Agreement:**  
   It will say, “Do you agree to the terms?”  
   Just click **“I agree”** (we all do this 😊).



2. **Install Location:**  
   It will ask where to install VS Code. You don’t need to change anything—just click **“Next”**.  
  

3. **Choose Tasks:**  
   You’ll see some checkboxes. Make sure these are checked:  
   - Add "Open with Code" to the right-click menu.  
   - Add to PATH (important!).  

   Then, click **“Next”** again.  


4. **Install:**  
   Now, click the **“Install”** button. Sit back and watch as VS Code gets installed on your computer. It might take a minute.  


---

### **5. Launch VS Code**  
When it’s done, you’ll see a button that says:  
**“Launch VS Code”**.  



Click it, and VS Code will open! 🎉 



### **6. Make Sure Everything Works**  
When VS Code opens, you’ll see a welcome screen. If it shows up, that means everything is working perfectly! 


### **Yay! You Did It! 🚀**  
Now you have VS Code on your computer. It’s like a big, colorful notebook where you can write all your Python magic. Ready to make some cool things? Let’s move on! 😊
